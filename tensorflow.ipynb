```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(-100)
```

```python
def softmax(x):
    return max(0,x)
```

```python
softmax(-10)
```

```python
softmax(100)
```

```python
def relu(x):
    return max(0.1*x,x)
```

```python
relu(10)
```

```python
relu(-100)
```

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(100)
```

```python
tanh(-300)
```
